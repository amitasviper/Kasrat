```python
import requests
import os

def download_file(url, filename=None):
    
    stream = requests.get(url, stream=True)
    
    if filename is None:
        filename = url.strip().split('/')[-1]
    
    with open(os.path.join('data', filename), 'wb') as f:
        for chunk in stream.iter_content(chunk_size=1024):
            if chunk:
                f.write(chunk)
    return filename
```
