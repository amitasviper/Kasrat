```python
def sigmoid_helper(val):
    return (1.0/(1+math.exp(val*-1)))

def sigmoid(val):
    ans = None
    if type(val) == type([]):
        ans = []
        for i in range(len(val)):
            ans.append(sigmoid_helper(val[i]))
    elif type(val) == type(np.array([])):
        ans = np.vectorize(sigmoid_helper)(val)
    elif type(val) == type(0) or type(val) == type(0.0):
        ans = sigmoid_helper(val)
    else:
        raise ValueError('Illegal argument type')
    return ans
```
